```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
import opendatasets as od
from torch.utils.data import DataLoader, TensorDataset
from scipy.io import loadmat

os.environ['KAGGLE_USERNAME'] = "Skieriya"
os.environ['KAGGLE_KEY'] = "2fb541a00a63ca348fa072108908245d"

dataset_url = "https://www.kaggle.com/datasets/scaomath/pde-dataset"
if os.path.exists("./pde-dataset") == False:
    od.download(dataset_url)

def load_pde_data(root_path, n_samples=100):
    for root, dirs, files in os.walk(root_path):
        for name in files:
            if "piececonst" in name:
                if name.endswith('.mat'):
                    data_path = root + "/" + name
                    raw_data = loadmat(data_path)

                    inputs = raw_data['coeff'][:n_samples]
                    outputs = raw_data['sol'][:n_samples]

                    inputs_tensor = torch.tensor(inputs, dtype=torch.float32)
                    outputs_tensor = torch.tensor(outputs, dtype=torch.float32)

                    return inputs_tensor.unsqueeze(3), outputs_tensor.unsqueeze(3)
    return None, None

X, Y = load_pde_data("./pde-dataset", 100)
dataset = TensorDataset(X, Y)
loader = DataLoader(dataset, batch_size=10, shuffle=True)

class SpectralConv2d(nn.Module):
    def __init__(self, in_channels, out_channels, modes1, modes2):
        super().__init__()
        self.modes1 = modes1
        self.modes2 = modes2


        self.weights1 = nn.Parameter(torch.view_as_complex(torch.rand(in_channels, out_channels, modes1, modes2, 2) * 0.02))
        self.weights2 = nn.Parameter(torch.view_as_complex(torch.rand(in_channels, out_channels, modes1, modes2, 2) * 0.02))

    def forward(self, x):
        batch_size = x.shape[0]
        x_ft = torch.fft.rfft2(x)

        out_ft = torch.zeros(batch_size, 32, x.size(2), x.size(3)//2 + 1, dtype=torch.cfloat, device=x.device)

        out_ft[:, :, :self.modes1, :self.modes2] = torch.einsum("bixy,ioxy->boxy",
                                                               x_ft[:, :, :self.modes1, :self.modes2],
                                                               self.weights1)

        out_ft[:, :, -self.modes1:, :self.modes2] = torch.einsum("bixy,ioxy->boxy",
                                                                x_ft[:, :, -self.modes1:, :self.modes2],
                                                                self.weights2)

        # Step 4: Move back to the Normal Spatial Domain (Images)
        return torch.fft.irfft2(out_ft, s=(x.size(2), x.size(3)))

class FNO2d(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc0 = nn.Linear(3, 32)

        self.conv0 = SpectralConv2d(32, 32, 12, 12)
        self.conv1 = SpectralConv2d(32, 32, 12, 12)
        self.conv2 = SpectralConv2d(32, 32, 12, 12)
        self.conv3 = SpectralConv2d(32, 32, 12, 12)

        self.w0 = nn.Conv2d(32, 32, 1)
        self.w1 = nn.Conv2d(32, 32, 1)
        self.w2 = nn.Conv2d(32, 32, 1)
        self.w3 = nn.Conv2d(32, 32, 1)

        self.fc1 = nn.Linear(32, 128)
        self.fc2 = nn.Linear(128, 1)

    def forward(self, x):
        batch_size = x.shape[0]
        res_x = x.shape[1]
        res_y = x.shape[2]

        grid_x = torch.linspace(0, 1, res_x)
        grid_y = torch.linspace(0, 1, res_y)
        grid_x, grid_y = torch.meshgrid(grid_x, grid_y, indexing='ij')
        grid = torch.stack((grid_x, grid_y), dim=2)
        grid = grid.unsqueeze(0).repeat(batch_size, 1, 1, 1).to(x.device)

        x = torch.cat((x, grid), dim=3)
        x = self.fc0(x)
        x = x.permute(0, 3, 1, 2)

        x1 = self.conv0(x)
        x2 = self.w0(x)
        x = F.gelu(x1 + x2)

        x1 = self.conv1(x)
        x2 = self.w1(x)
        x = F.gelu(x1 + x2)

        x1 = self.conv2(x)
        x2 = self.w2(x)
        x = F.gelu(x1 + x2)

        x1 = self.conv3(x)
        x2 = self.w3(x)
        x = F.gelu(x1 + x2)

        x = x.permute(0, 2, 3, 1)
        x = self.fc1(x)
        x = F.gelu(x)
        x = self.fc2(x)
        return x

device = torch.device('cpu')
if torch.cuda.is_available():
    device = torch.device('cuda')

model = FNO2d().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

print("Starting training")
for epoch in range(15):
    model.train()
    for batch in loader:
        batch_x = batch[0].to(device)
        batch_y = batch[1].to(device)

        optimizer.zero_grad()
        prediction = model(batch_x)

        diff = (prediction - batch_y)**2
        mse_loss = torch.mean(diff)
        norm = torch.mean(batch_y**2) + 0.00000001
        loss = mse_loss / norm

        loss.backward()
        optimizer.step()

def check_for_errors(model, input_data, target_data, label):
    model.eval()
    input_data = input_data.to(device)
    prediction = model(input_data)
    prediction = prediction.cpu()

    error_map = torch.abs(prediction - target_data)
    max_error = torch.max(error_map).item()

    status = "NORMAL"
    if max_error > 0.05:
        status = "ANOMALY"

    print("Sample: " + label + " Max Error: " + str(max_error) + " Status: " + status)
    return prediction, error_map

test_input = X[10:11]
test_target = Y[10:11]

broken_input = test_input.clone()
broken_input[0, 150:250, 150:250, 0] = 2.0

pred1, err1 = check_for_errors(model, test_input, test_target, "Normal")
pred2, err2 = check_for_errors(model, broken_input, test_target, "Corrupted")

plt.figure(figsize=(15, 6))

plt.subplot(2, 3, 1)
plt.imshow(test_input[0, :, :, 0], cmap='jet')
plt.title("Normal Input")

plt.subplot(2, 3, 2)
plt.imshow(pred1[0, :, :, 0].detach().numpy(), cmap='jet')
plt.title("Prediction")

plt.subplot(2, 3, 3)
plt.imshow(err1[0, :, :, 0].detach().numpy(), cmap='hot')
plt.title("Error Map")

plt.subplot(2, 3, 4)
plt.imshow(broken_input[0, :, :, 0], cmap='jet')
plt.title("Corrupted Input")

plt.subplot(2, 3, 5)
plt.imshow(pred2[0, :, :, 0].detach().numpy(), cmap='jet')
plt.title("Confused Prediction")

plt.subplot(2, 3, 6)
plt.imshow(err2[0, :, :, 0].detach().numpy(), cmap='hot')
plt.title("High Error Map")

plt.show()
```

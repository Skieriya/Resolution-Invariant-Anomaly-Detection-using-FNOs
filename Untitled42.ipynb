```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import scipy.ndimage as ndimage


# 1. FNO Architecture


class SpectralConv2d(nn.Module):
    def __init__(self, in_channels, out_channels, modes1, modes2):
        super(SpectralConv2d, self).__init__()
        self.in_channels = in_channels
        self.out_channels = out_channels
        self.modes1 = modes1
        self.modes2 = modes2
        self.scale = 1 / (in_channels * out_channels)
        self.weights1 = nn.Parameter(self.scale * torch.rand(in_channels, out_channels, self.modes1, self.modes2, dtype=torch.cfloat))
        self.weights2 = nn.Parameter(self.scale * torch.rand(in_channels, out_channels, self.modes1, self.modes2, dtype=torch.cfloat))

    def compl_mul2d(self, input, weights):
        return torch.einsum("bixy,ioxy->boxy", input, weights)

    def forward(self, x):
        batchsize = x.shape[0]
        x_ft = torch.fft.rfft2(x)
        out_ft = torch.zeros(batchsize, self.out_channels, x.size(-2), x.size(-1)//2 + 1, dtype=torch.cfloat, device=x.device)
        out_ft[:, :, :self.modes1, :self.modes2] = self.compl_mul2d(x_ft[:, :, :self.modes1, :self.modes2], self.weights1)
        out_ft[:, :, -self.modes1:, :self.modes2] = self.compl_mul2d(x_ft[:, :, -self.modes1:, :self.modes2], self.weights2)
        x = torch.fft.irfft2(out_ft, s=(x.size(-2), x.size(-1)))
        return x

class FNO2d(nn.Module):
    def __init__(self, modes1, modes2, width):
        super(FNO2d, self).__init__()
        self.modes1 = modes1
        self.modes2 = modes2
        self.width = width
        self.fc0 = nn.Linear(3, self.width) # Input: (u, grid_x, grid_y) -> 3 channels

        self.conv0 = SpectralConv2d(self.width, self.width, self.modes1, self.modes2)
        self.conv1 = SpectralConv2d(self.width, self.width, self.modes1, self.modes2)
        self.conv2 = SpectralConv2d(self.width, self.width, self.modes1, self.modes2)
        self.conv3 = SpectralConv2d(self.width, self.width, self.modes1, self.modes2)

        self.w0 = nn.Conv2d(self.width, self.width, 1)
        self.w1 = nn.Conv2d(self.width, self.width, 1)
        self.w2 = nn.Conv2d(self.width, self.width, 1)
        self.w3 = nn.Conv2d(self.width, self.width, 1)

        self.fc1 = nn.Linear(self.width, 128)
        self.fc2 = nn.Linear(128, 1)

    def forward(self, x):
        grid = self.get_grid(x.shape, x.device)
        x = torch.cat((x, grid), dim=-1)

        x = self.fc0(x)
        x = x.permute(0, 3, 1, 2)

        x1 = self.conv0(x)
        x2 = self.w0(x)
        x = x1 + x2
        x = F.gelu(x)

        x1 = self.conv1(x)
        x2 = self.w1(x)
        x = x1 + x2
        x = F.gelu(x)

        x1 = self.conv2(x)
        x2 = self.w2(x)
        x = x1 + x2
        x = F.gelu(x)

        x1 = self.conv3(x)
        x2 = self.w3(x)
        x = x1 + x2

        x = x.permute(0, 2, 3, 1)
        x = self.fc1(x)
        x = F.gelu(x)
        x = self.fc2(x)
        return x

    def get_grid(self, shape, device):
        batchsize, size_x, size_y = shape[0], shape[1], shape[2]
        gridx = torch.linspace(0, 1, size_x, dtype=torch.float32, device=device)
        gridy = torch.linspace(0, 1, size_y, dtype=torch.float32, device=device)
        meshx, meshy = torch.meshgrid(gridx, gridy, indexing='ij')
        meshx = meshx.unsqueeze(0).repeat(batchsize, 1, 1)
        meshy = meshy.unsqueeze(0).repeat(batchsize, 1, 1)
        grid = torch.stack((meshx, meshy), dim=-1)
        return grid


# 2. Data Generation (Physics Simulation)


def generate_data(num_samples, resolution=64, anomaly_ratio=0.0):
    inputs = []
    outputs = []
    labels = []

    for i in tqdm(range(num_samples), desc=f"Generating {num_samples} samples"):
        # Random Heat Source
        field = np.zeros((resolution, resolution))
        num_blobs = np.random.randint(1, 4)
        for _ in range(num_blobs):
            x0, y0 = np.random.randint(0, resolution, 2)
            sigma = np.random.uniform(3, 8)
            X, Y = np.meshgrid(np.arange(resolution), np.arange(resolution))
            blob = np.exp(-((X-x0)**2 + (Y-y0)**2) / (2*sigma**2))
            field += blob

        # Normal Physics: Diffusion (Gaussian Blur)
        evolved_normal = ndimage.gaussian_filter(field, sigma=1.0)

        is_anomaly = np.random.rand() < anomaly_ratio

        if is_anomaly:
            # Inject Anomaly
            if np.random.rand() > 0.5:
                # Hotspot: Physics violation (energy spike)
                cx, cy = resolution//2, resolution//2
                evolved_normal[cx-3:cx+3, cy-3:cy+3] += 2.0
            else:
                # Blockage: Physics violation (energy loss)
                evolved_normal[10:20, 10:20] = 0
            labels.append(1)
            outputs.append(evolved_normal)
        else:
            labels.append(0)
            outputs.append(evolved_normal)

        inputs.append(field)

    inputs = np.array(inputs)[..., np.newaxis]
    outputs = np.array(outputs)[..., np.newaxis]

    return torch.tensor(inputs, dtype=torch.float32), torch.tensor(outputs, dtype=torch.float32), torch.tensor(labels)

# 3. Main


device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# 1. Generate Data
print("\n[1/4] Generating Training Data (Normal Physics)...")
X_train, y_train, _ = generate_data(1000, anomaly_ratio=0.0)

print("\n[2/4] Generating Test Data (Mixed Normal & Anomaly)...")
X_test, y_test, labels_test = generate_data(500, anomaly_ratio=0.3)

# 2. Train Model
model = FNO2d(modes1=12, modes2=12, width=32).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
train_loader = torch.utils.data.DataLoader(torch.utils.data.TensorDataset(X_train, y_train), batch_size=16, shuffle=True)

print("\n[3/4] Training FNO to learn Normal Physics...")
epochs = 20
for ep in range(epochs):
    model.train()
    total_loss = 0
    for x, y in train_loader:
        x, y = x.to(device), y.to(device)
        optimizer.zero_grad()
        out = model(x)
        loss = F.mse_loss(out, y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    if (ep+1) % 5 == 0:
        print(f"Epoch {ep+1}/{epochs}, Loss: {total_loss/len(train_loader):.6f}")

# 3. Anomaly Detection
print("\n[4/4] Running Anomaly Detection & Visualization...")
model.eval()
errors = []
preds = []

with torch.no_grad():
    for i in range(len(X_test)):
        x = X_test[i].unsqueeze(0).to(device)
        y_true = y_test[i].unsqueeze(0).to(device)
        y_pred = model(x)

        error = torch.norm(y_pred - y_true).item()
        errors.append(error)
        preds.append(y_pred.squeeze().cpu().numpy())

mean_err = np.mean(errors)
std_err = np.std(errors)
threshold = mean_err + 1.5 * std_err
print(f"Detection Threshold: {threshold:.4f}")

# Plotting
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
indices = [0, 15, 30]
for idx, val in enumerate(indices):
    # Input
    axs[0, idx].imshow(X_test[val].squeeze(), cmap='hot')
    axs[0, idx].set_title(f"Input (Sample {val})")

    # Error Map
    diff = np.abs(preds[val] - y_test[val].squeeze().numpy())
    is_anomaly = labels_test[val].item() == 1
    status = "Anomaly" if is_anomaly else "Normal"

    axs[1, idx].imshow(diff, cmap='viridis')
    axs[1, idx].set_title(f"Error Map\nGT: {status}\nError: {errors[val]:.2f}")

plt.tight_layout()
plt.show()

# Calculate Accuracy
detected = [1 if e > threshold else 0 for e in errors]
correct = sum([1 for d, l in zip(detected, labels_test.numpy()) if d == l])
print(f"Detection Accuracy: {correct}/{len(labels_test)} ({correct/len(labels_test)*100:.2f}%)")


# 4. Zero-Shot Super Resolution Test

print("\n[BONUS] Testing Zero-Shot Super Resolution (128x128)...")
X_high, y_high, _ = generate_data(1, resolution=128, anomaly_ratio=0.0)

with torch.no_grad():
    x_dev = X_high.to(device)
    y_pred_high = model(x_dev).cpu()

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(X_high.squeeze(), cmap='hot')
axs[0].set_title("High Res Input (128x128)")
axs[1].imshow(y_pred_high.squeeze(), cmap='hot')
axs[1].set_title("FNO Prediction (Trained on 64x64)")
axs[2].imshow(y_high.squeeze(), cmap='hot')
axs[2].set_title("Ground Truth")
plt.show()
```
